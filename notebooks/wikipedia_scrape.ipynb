```python
Country = []
for c in Countries:
    Country.append(c[1])
```

```python
date = []
country = []
about = []

flag = False
stop = False

for t in text:
    if stop:
        break
        
    # break on the last sentence we want   
    if("The United States confirmed its first death, a man from Washington near the Seattle area" in t):
        stop = True
    
    if ("WHO Situation Report" in t):
        
        day = int(t[21:23])-11
        flag = True
    
    if(flag):
        if("Situation Report" in t or "[edit]" in t):
            continue
        else:
            sentence = t.split("[") 
            
            if day<10:
 #               date.append("0"+str(day)+"-02-2020")
                d = "0"+str(day)+"-02-2020"
            else:
#                date.append(str(day)+"-02-2020")
                d = str(day)+"-02-2020"
            
        
            for c in Country:
                if c in sentence[0]:
                    if(len(sentence[0])<10):
                        break
                    about.append(sentence[0])
                    country.append(c)
                    date.append(d)
                    break
    
        
        
```
